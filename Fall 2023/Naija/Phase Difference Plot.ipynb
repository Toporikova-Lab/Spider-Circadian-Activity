# Phase Difference Plot

```python
#Import relevant packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
from datetime import timedelta
import seaborn as sns
sns.set()
```

```python
df1 = pd.read_csv("Practice 2_data1.csv")
df1["Time"] = pd.to_datetime(df1["Time"])
df1 = df1.set_index("Time")
```

```python
roll = df1.rolling(window = "90min", axis = 0).mean()
roll["s2"].plot(color='b', label= "Spider 2", title = "1.5 hr Moving Average", legend = True, ylabel = "Activity (Crossings)")
roll["s5"].plot(color='g', label= "Spider 5", legend = True)
```

```python
roll["s6"].plot(color='r', label= "Spider 6", title = "1.5 hr Moving Average", legend = True, ylabel = "Activity (Crossings)")
roll["s5"].plot(color='g', label= "Spider 5", legend = True)
```
