# Noise Reduction

## Noise Reduction: DD data

We want to perform noise reduction on the data frame of DD binarized data. We set the date_time as the index and only include the spider columns.

```python
import numpy as np
import pandas as pd
pd.options.mode.chained_assignment = None

file_names = "Metazygia wittfeldae Monitor 1 activity"

df_DD = pd.read_csv(file_names + "_DD_binary.csv", index_col = 0)


df_DD["Date_Time"] = pd.to_datetime(df_DD.Date + ' ' + df_DD.Time)
df_DD = df_DD.set_index("Date_Time")
df_DD = df_DD.loc[:, df_DD.columns.str.startswith('s')]
display(df_DD)
```

This function loops through the dataframe columns and performs noise reduction (replaces 0 surrounded by 1s with values of 1) and saves it as a csv file.

```python
col_DD = list(df_DD)
for x in col_DD:
    df_DD[1:].loc[(df_DD[x].shift(-1) == 1) == (df_DD[x].shift(1) == 1), x] = 1
    
df_DD.to_csv(file_names + "_DD_binary_noiseReduction.csv")
```
