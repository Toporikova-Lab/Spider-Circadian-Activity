# Importing

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

from sklearn import preprocessing

df = pd.read_csv("Metazygia wittfeldae Monitor 1 activity_DD.csv")
df
```

# Setting up Binarizer

```python
from sklearn.preprocessing import Binarizer

bin = Binarizer()
```

# Extracting Columns and Binarizing

```python
for x in df.columns[4:]:
    col = df[x].values
    col = col.reshape(-1, 1)
    df[x] = bin.fit_transform(col)

df.to_csv("Metazygia wittfeldae Monitor 1 activity_DD_binary.csv")
display(df)
```
