```python
#Import Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
from datetime import timedelta

#Read Spider activity CSV file
df = pd.read_csv('Metazygia wittfeldae Monitor 1 activity_LD_binary.csv', index_col = 0)

#Cut out certain columns to make data more managable to test on
#df.drop(df.iloc[:, 6:], axis = 1, inplace = True)
df1 = df.copy()
df1 = df1[['Date_Time', 's10']]
#df1.drop(df1.index[30:], inplace = True)

df2 = df.copy()
df2 = df2[['Date_Time', 's32']]
#df2.drop(df2.index[61:], inplace = True)

display(df1)
display(df2)
```

Create switch column

```python
df_switch1 = df1.copy()

df_switch1['switch'] = df_switch1['s10'].diff()
from_1_to_0 = df_switch1[df_switch1['switch'] == -1]
from_0_to_1 = df_switch1[df_switch1['switch'] == 1]

display(df_switch1)

df_switch2 = df2.copy()

df_switch2['switch'] = df_switch2['s32'].diff()
from_1_to_0 = df_switch2[df_switch2['switch'] == -1]
from_0_to_1 = df_switch2[df_switch2['switch'] == 1]

display(df_switch2)
```

Find start and end times

```python
#print(df_switch[df_switch['switch'] == '1.0'].index.values)

runStart1 = df_switch1.index[df_switch1['switch'] == 1].tolist()
runEnd1 = df_switch1.index[df_switch1['switch'] == -1].tolist()

runStart2 = df_switch2.index[df_switch2['switch'] == 1].tolist()
runEnd2 = df_switch2.index[df_switch2['switch'] == -1].tolist()
```

Find run duration through subtraction

```python
runDuration1 = np.subtract(runEnd1, runStart1)

runDuration2 = np.subtract(runEnd2, runStart2)
```

```python
new_df1 = pd.DataFrame(runDuration1, columns = ['s10'])
display (new_df1)

new_df2 = pd.DataFrame(runDuration2, columns = ['s32'])
display (new_df2)
```

Combining dataframes

```python
df3 = pd.concat([new_df1, new_df2.set_index(new_df1.index[-len(new_df2):])], axis = 1)

display(df3)
```

```python
if df_switch1.at[0, 's31'] == 1:
    df_switch1.at[0, 'switch'] = 1
    
if df_switch1.at[df_switch1.index[-1], 's31'] == 1 and df_switch1.at[df_switch1.index[-2], 's31'] == 1:
    df_switch1.at[df_switch1.index[-1], 'switch'] = -1
    
if df_switch1.at[df_switch1.index[-1], 's31'] == 1 and df_switch1.at[df_switch1.index[-2], 's31'] == 0:
    df_switch1.at[df_switch1.index[-1], 'switch'] = 0

runStart1 = df_switch1.index[df_switch1['switch'] == 1].tolist()
runEnd1 = df_switch1.index[df_switch1['switch'] == -1].tolist()

print(len(runStart1))
print(len(runEnd1))

runDuration1 = np.subtract(runEnd1, runStart1)
print(runDuration1)
```

# New Dataframe

Making a new dataframe with all run durations for all spiders. Currently our error is that the maximum rows is set to the run duration of s1, however there are spiders that have longer duration counts than s1.

```python
newdf = pd.DataFrame()
df_switch = df1.copy()

for x in df1.loc[:, df1.columns.str.startswith('s')]:
    df_switch['switch'] = df_switch[x].diff()
    from_1_to_0 = df_switch[df_switch['switch'] == -1]
    from_0_to_1 = df_switch[df_switch['switch'] == 1]
           
    if df_switch.at[0, x] == 1:
        df_switch.at[0, 'switch'] = 1
    
    if df_switch.at[df_switch.index[-1], x] == 1 and df_switch.at[df_switch.index[-2], x] == 1:
        df_switch.at[df_switch.index[-1], 'switch'] = -1
    
    if df_switch.at[df_switch.index[-1], x] == 1 and df_switch.at[df_switch.index[-2], x] == 0:
        df_switch.at[df_switch.index[-1], 'switch'] = 0       
    
    runStart = df_switch.index[df_switch['switch'] == 1].tolist()
    runEnd = df_switch.index[df_switch['switch'] == -1].tolist()
    runDuration = np.subtract(runEnd, runStart)

    newdf.loc[:, x] = pd.Series(runDuration)
    #newdf.loc[:, x] = runDuration.tolist()
    
display(newdf)
```
