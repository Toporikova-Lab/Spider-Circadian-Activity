# Create Histogram

Creates histogram of all spider run durations with fixed axes added manually. 

```python
def draw_histogram(col):    
    plt.title(col + ' DD Run Duration Distribution')
    plt.xlabel('Run Duration')
    plt.ylabel('Count')
    #maxDuration = df1.max().max()
    #maxCounts = df1.idxmax()
    #maxCount = maxCounts.max()
    plt.grid(True)
    plt.xlim(0, 50)
    plt.ylim(0, 500)
    #plt.xlim(0, maxDuration + 10)
    #plt.ylim(0, maxCount * 1.5)
```

# Show Histograms

Shows all spider run duration histograms

```python
def show_all_histograms(df):
    for i, col in enumerate(df.columns):
        plt.figure(i)
        plt.hist(df[col])
    
        draw_histogram(col)
        
        plt.savefig(col + ".jpeg")
        
show_all_histograms(df)
```
