```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
from datetime import timedelta

#Read Spider activity CSV file
df = pd.read_csv('Metazygia wittfeldae Monitor 1 activity_LD_binary.csv', index_col = 0)
df1 = df.copy()
display(df1)
```

Creates switch column that identifies where spider started and stopped moving with error checking. Then creates two lists; runStart is a list with the indices of when the spider started moving, runEnd is a list with the indices of when the spider stopped moving. Run duration is found by subtracting these two lists. Run duration of each spider is then added to a new dataframe. If a spider has a shorter run duration counts than another spider, the empty rows are filled with NaN.

```python
def allSpiderRuns(df1):
    newdf = pd.DataFrame()
    df_switch = df1.copy()
    durationList = []
    maxDuration = []
    
    for x in df1.loc[:, df1.columns.str.startswith('s')]:
        df_switch['switch'] = df_switch[x].diff()
        from_1_to_0 = df_switch[df_switch['switch'] == -1]
        from_0_to_1 = df_switch[df_switch['switch'] == 1]
           
        if df_switch.at[0, x] == 1:
            df_switch.at[0, 'switch'] = 1
    
        if df_switch.at[df_switch.index[-1], x] == 1 and df_switch.at[df_switch.index[-2], x] == 1:
            df_switch.at[df_switch.index[-1], 'switch'] = -1
    
        if df_switch.at[df_switch.index[-1], x] == 1 and df_switch.at[df_switch.index[-2], x] == 0:
            df_switch.at[df_switch.index[-1], 'switch'] = 0       
    
        runStart = df_switch.index[df_switch['switch'] == 1].tolist()
        runEnd = df_switch.index[df_switch['switch'] == -1].tolist()
        runDuration = np.subtract(runEnd, runStart)
        runDurationLength = len(runDuration)
    
        durationList.append(runDurationLength)
    
        newdf.loc[:, x] = pd.Series(runDuration) 

        for i in newdf.loc[:, newdf.columns.str.startswith('s')]:  
            maxRun = max(durationList) - 1
            newdf = newdf.reindex(list(range(0, maxRun))).reset_index(drop = True)    
            newdf.loc[:, x] = pd.Series(runDuration)
    
    display(newdf)
    newdf.to_csv("Metazygia wittfeldae Monitor 1 activity_LD_run_duration.csv")
    
allSpiderRuns(df1)
```
