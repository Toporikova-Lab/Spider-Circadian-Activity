```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%run plotting_functions.ipynb
```

```python
# !!! ONLY RUN THIS FUNCTION ON NON-MASKED DATA SETS !!!
def average_all_dfs(file_names):
    collected_df = pd.DataFrame(columns = ['Day', 'Time', 'Light'])

    final_sum = np.empty(1440)
    
    for n, file_name in enumerate(file_names):
        df = process_file(file_name)

        light_on_times = df.index[(df['Light'] == 1) & (df['Light'].shift(1) == 0)]
        first_on = light_on_times[0]

        df.drop(['Light', 'Day', 'Time'], axis=1, inplace=True)
        combined = df.mean(axis=1)

        for on_time in light_on_times:
            section = combined[on_time : on_time + 1440]
            final_sum[:len(section)] += np.array(section)

    return final_sum
```

```python
file_names = ['../Behavioral/Data/StA LD 1205-1212-2023.txt', '../Behavioral/Data/StA LD 1117-1128-2023.txt', '../Behavioral/Data/StB LD-DD 1212-1224-2023.txt']
file_names_2 = ['../Behavioral/Data/StA LD 1205-1212-2023.txt', '../Behavioral/Data/StA LD 1117-1128-2023.txt']
fig, axs = plt.subplots(2, 1)

control_avg_1 = average_all_dfs(file_names)
control_avg_2 = average_all_dfs(file_names_2)

axs[0].plot(control_avg_1)
axs[1].plot(control_avg_2)
print(control_avg_1 - control_avg_2)
```
