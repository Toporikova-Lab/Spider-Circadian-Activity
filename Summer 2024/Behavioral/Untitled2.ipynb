```python
import numpy as np
from scipy.fft import fft, ifft
import matplotlib.pyplot as plt
```

```python
N = 600
# Sample spacing
T = 1.0 / 800.0
x = np.linspace(0.0, N*T, N)
y = 2*np.sin(20.0 * 2*np.pi*x) + 0.5*np.sin(50.0 * 2.0*np.pi*x)

yf = fft(y)
xf = np.linspace(0.0, 1.0/(2.0*T), N//2)

plt.subplot(2, 1, 1)
plt.plot(x, y)
plt.title('Signal')

plt.subplot(2, 1, 2)
plt.plot(xf, 2.0/N * np.abs(yf[:N//2]))
#plt.title('FFT Coefficients')
plt.show()
```
