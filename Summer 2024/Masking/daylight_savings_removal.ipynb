```python
def time_mapper(timestr, index, jump):
    if index > jump:
        time = pd.to_datetime(timestr, format='%H:%M:%S')
        return (time - pd.Timedelta(1, unit='hours')).strftime('%H:%M:%S')
    return timestr

def date_mapper(datestr, timestr, index, jump):
    if index > jump:
        time = pd.to_datetime(timestr, format='%H:%M:%S')
        date = pd.to_datetime(datestr, format='%d %b %y')
        if time.hour == 23:
            return (date - pd.Timedelta(1, unit='days')).strftime('%d %b %y')
    return datestr

def remove_daylight_savings(in_name, out_name):
    df = pd.read_csv(in_name, sep='\t', header=None)
    display(df)
    
    times = pd.to_datetime(df[2], format='%H:%M:%S')
    jump_index = df.index[(times - times.shift(1)).dt.seconds / 3600 > 1][0]
    
    df[2] = df.apply(lambda x: time_mapper(x[2], x[0], jump_index), axis=1)
    df[1] = df.apply(lambda x: date_mapper(x[1], x[2], x[0], jump_index), axis=1)
    
    df.to_csv(out_name, sep='\t', header=False, index=False)
```

```python
remove_daylight_savings('Steatoda A masking 10 am.txt', 'Steatoda A masking 10 am - extra hour removed.txt')
```
