# Raster plots

## Binary raster plot

First, process data from a file using the process_file function into a pandas DataFrame.
Then use the raster_binary function to plot the binarized activity of the spider Sp9, with the title "MsA 9".

```python
df = process_file('Data/MsA DD-LD 0624-0708-2024.txt')

_ =raster_binary(df, 'Sp9', 'MsA 9')
```

## Line raster plot

This works the same as the binary raster plot, but plots a line graph instead of binarized data.

```python
df = process_file('Data/MsA DD-LD 0624-0708-2024.txt')

_ = raster_line(df, 'Sp9', 'MsA 9')
```

## Rolling averages

Both the binary and line graph raster plots have an optional rolling_window parameter (in minutes). Changing this value to be greater than 1 will apply a rolling average across the plotted data. For the line graph, this has the effect of smoothing out the line; for the binary data, this has the effect of spreading out the bars to make shorter periods of activity more visible.

```python
df = process_file('Data/MsA DD-LD 0624-0708-2024.txt')

_ = raster_binary(df, 'Sp9', 'MsA 9 with rolling average', rolling_window=6)
_ = raster_line(df, 'Sp9', 'MsA 9 with rolling average', rolling_window=12)
```
