```python
import os
import numpy as np
%matplotlib inline
from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
zero_mean_series = np.random.normal(loc=0.0, scale=1., size=100)
```

```python
plt.figure(figsize=(5.5, 5.5))
length = (len(zero_mean_series))
plt.plot((range(0,length)), zero_mean_series)
plt.show()
```

```python
random_walk = np.cumsum(zero_mean_series)
```

```python
plt.figure(figsize=(5.5, 5.5))
length_walk = (len(random_walk))
plt.plot((range(0,length_walk)),random_walk)
plt.show()
```
