##### Since the inital breakthroughs in chronobiology in the 1950s, scientists have been studying the interal clocks of organisms from a wide variety of species and sizes (Dunlap, Jac C., et al 11). Notably, researchers have spent significant time studying the internal biological rhythms of insects and sometimes spiders. Colin Pittendrigh was a successful biologist who first started researching the circadian rhythms of Drosophila, or fruit flies, which led to a greater understanding of insect chronobiology (Tataroglu 2015). Many processes of the life cycle of Drosophila, such as eclosion, courtship, and foraging, demonstrate the existence of an internal oscillator in the flies (Maruo 2007). Pittendrigh determind that both temperature and certain wavelengths of light affected the circadian rhythm of the flies, shifting their internal clocks (Tataroglu 2015). Beyond fruit flies, biologists have determined that the majority of living organisms from the smallest cyanobacteria to humans have circadian rhythms that elicit certain behavioral repsonses at specific periods of the time (Tataroglu 2015). 
##### These processes can be examined and tested through the usage of a specific appartus designed to measure the locomotive acitivty of small organisms. Trikinetic's LAM25 monitor is often employed by scientists when trying to measure this acitivty. The apparatus works by firing infared lasers into a tube with a spider and a given amount of medium to supply nutrietns and water to the subject, and the organism is left in the tube for a few days. Whenever the insect or spider crosses the laser, it triggers an event which is recorded in a connected computer. The events are stored and can be graphed to visualize at what times the subject is most active and therefore when the subject's circadian rhythm fosters locomotive acivity. The amount of light shone in the room can be adjusted to simulate different cycles of day and night. Similarly, the temperature of the room can be changed to experiment with the temperature's effects on the internal clocks of the insects or spiders. Below is an example of what an ideal visualization of data looks like using spiders.

```python
df2 = pd.read_csv('Practice 1_data2.csv')
#plt.plot(df2.Time, df2.s2, label = 's2')
plt.plot(df2.Time, df2.s5, label = 's5')
plt.plot(df2.Time, df2.s6, label = 's6')
plt.xticks(np.arange(0, len(df2)+1, 60))
plt.title('Week 1 Spider Crossings')
plt.xlabel('Time of Day')
plt.ylabel('Number of Crossings')
plt.legend()
plt.show()
```

#### The graph displays the number of triggered events, or "crossings," in a given period of time, with the y-axis denoting the number of crossings and the x-axis displaying the time. This data can be analyzed to determine when the spiders are most active. Any time there is a recorded crossing, the spider is active. For example, s6 has a spike in its crossings between 14:25:00 and 15:25:00, highlighting a period of activity for the spider. Contrastingly, s6 has no crossings at all after 15:25:00, implying that the spider was likely resting during this time period. Using these periods of activity and inactivty, one can determine roughly the timing of the internal clock of the spider. This method also allows one to compare and contrast a range of spiders and their periods of activity. For instance, the two spiders displayed above have different but similar periods of activity, displaying that their circadian rhythms are relatively in sync during this window of time. However, there are small periods of time where one spider is active and the other is not. For example, in the window beteen 14:00:00 and 14:25:00, s5 is acitve while s6 does not cross at all. Despite being a relatively short amount of time, this still displays an observable disparity between the rythms of the two spiders. Although more time is needed to draw further conclusions from this case, the data visualization is still effective. The longer the time recorded, the more a pattern will be displayed.
