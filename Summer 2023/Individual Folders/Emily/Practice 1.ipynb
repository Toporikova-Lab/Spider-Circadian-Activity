Scientists are trying to conclude how insect circadian rhythms adapt to different environments. Currently the only indicators that affect the activity of spiders and insects are courting routines, searching for food, and exploring their surroundings (Mauro 2007). Previous research indicates that insects are able to adapt when their normal light and dark cycle is disrupted (Chiu 2010). Insects exposed to normal light and dark cycles had similar energy spikes in the morning and evening as insects who only experience dark cycles (Chiu 2010). It was also found that insects held under constant temperature and stimulus can maintain the same movement cycle but insects bombarded with stimuli (light) at irregular times will experience a shift in their cycle (Tataroglu 2015). If spiders react similarly to the Drosophilia fly, then the spiders will continue to have an increase in energy at the same times each day.

Activity levels of spiders and insects can be monitored by a LAM25 monitor. This machine holds the insects in individual capsules and shines an infared beam down the middle of the capusule. The researcher will place food on one side and an environment for resting on the other side. Activity is measured by counting how many times the insect disrupts the infared beam. This data is exported to a data sheet with time stamps to better understand when the insects are moving. The data collected will also show high and low levels of activity throughout the day which can be compared across individuals to gain a better understanding of how envirnments affect insect circadian rhythms.

```python
df2=pd.read_csv('Practice 1_data2.csv', usecols=['Time',"s5", "s6"])
df2.plot( x='Time',linestyle='-', marker='.', figsize=(20,10))
plt.xlabel('Time')
plt.ylabel('Number of Beam Crossings')
plt.title('Activity of Spiders Over Time (Fig. 1)')
plt.show()
```

```python
df12=df2.iloc[0:35]
df13=df2.iloc[35:98]
df14=df2.iloc[98:159]
df15=df2.iloc[159:218]
df16=df2.iloc[218:278]
df17=df2.iloc[278:338]

df12s5=df12['s5']
df12s6=df12['s6']
df13s5=df13['s5']
df13s6=df13['s6']
df14s5=df14['s5']
df14s6=df14['s6']
df15s5=df15['s5']
df15s6=df15['s6']
df16s5=df16['s5']
df16s6=df16['s6']
df17s5=df17['s5']
df17s6=df17['s6']

num=range(0,6)
means5=[df12s5, df13s5, df14s5, df15s5, df16s5, df17s5]
ms=np.empty(6)
for i in num:
    a=np.mean(means5[i])
    ms[i]=a
    
means6=[df12s6, df13s6, df14s6, df15s6, df16s6, df17s6]
ms6=np.empty(6)
for i in num:
    a=np.mean(means6[i])
    ms6[i]=a
    
ind = np.arange(6)


plt.figure(figsize=(10,5))

width = .3     

plt.bar(ind , ms , width, label='s5')
plt.bar(ind + width, ms6, width, label='s6')
plt.legend()
plt.xticks([0,1,2,3,4,5],['12:00', '13:00', '14:00', '15:00', '16:00', '17:00'])
plt.xlabel('Time')
plt.ylabel('Average Movement')
plt.title('Spiders s5 and s6 Average Movement Each Hour (Fig.2)')
```
