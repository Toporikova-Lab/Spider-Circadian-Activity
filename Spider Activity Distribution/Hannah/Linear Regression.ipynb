```python
import pandas as pd
df_boston = pd.read_csv('House Prices.csv') #df for linear regression tutorial (can be modified with spider df)
df_boston
```

```python
import statsmodels.api as sm
```

```python
y = df_boston['Value'] # dependent variable
x = df_boston['Rooms'] # independent variable
```

```python
x = sm.add_constant(x) # adding a constant
lm = sm.OLS(y,x).fit() # fitting the model
```

```python
lm.predict(x)
```

```python
lm.summary()
```

```python
# Rooms coef: 9.1021
# Constant coef: - 34.6706
# Linear equation: ùë¶ = ùëéùë• + ùëè
y_pred = 9.1021 * x['Rooms'] - 34.6706
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()
# plotting the data points
sns.scatterplot(x=x['Rooms'], y=y)
#plotting the line
sns.lineplot(x=x['Rooms'],y=y_pred, color='red')
#axes
plt.xlim(0)
plt.ylim(0)
plt.show()
```
