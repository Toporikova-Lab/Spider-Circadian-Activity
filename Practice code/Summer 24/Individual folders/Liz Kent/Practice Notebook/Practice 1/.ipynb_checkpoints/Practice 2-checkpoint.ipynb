## Experiment 1: Modeling spider activity with white noise

In the [first chapter of the Time series analysis book](https://learning.oreilly.com/library/view/time-series-analysis/9780470272848/08_chap01.html) we learned that the easiest model for a time series is a white noise. That is the series of random events with the same mean and variance as the data. Let's try to model our spider data with a white noise. Such experiment provides a good visual test of hypothesis if spider behavior is random or intentional. If random, the spider activity will be similar to a random noise. However, if there are some  pattern associated with spider activity, the spider behavior should be quite different than noise

### Calculate mean and standard deviation of time series for each spider

Calculate mean and standard deviation of time series for each spider. Save as a dataframe **df_summary** which has spider id in the rows and mean(mu) and standard deviation (sigma) in columns 

```python
spider_df = ['s2','s5','s6']
mu = np.empty(3)
sigma = np.empty(3)
num = range(0,3)
for i in num:
    #calculate mean and store it 
    mean = np.mean(df[spider_df[i]])
    mu[i] = mean
    #calculate std and store it 
    std = np.std(df[spider_df[i]])
    sigma[i] = std
df_summary = pd.DataFrame(data = [mu, sigma])
#transpose of df to create correct columns
df_summary = df_summary.T
df_summary.columns = ['mu', 'sigma']
df_summary.index = spider_df
display(df_summary)
```

### Generate random time series using spiders mean and standard deviation

Now make a new data frame df_random, which is a copy of spider activity in  Week2_data1.csv. Replace every entry of spider activity with a random number with the same mean and standard deviation as spider activity. Plot spider activity and white noise on the same plot.

```python
df = pd.read_csv('Practice 2_data1.csv')
df_random = df 

for i in range(0, df.shape[0]):
    a = abs(int(np.random.normal(df_summary.iloc[0,0], df_summary.iloc[0,1])))
    df_random.iloc[i,1] = a
    b = abs(int(np.random.normal(df_summary.iloc[1,0], df_summary.iloc[1,1])))
    df_random.iloc[i,2] = b  
    c = abs(int(np.random.normal(df_summary.iloc[2,0], df_summary.iloc[2,1])))
    df_random.iloc[i,3] = c
display(df_random)
```

##### Entrainment is a property of circadian systems where the endogenous cycle of an organism is synchronized to a period of external stimuli (Duffy, Wright, 2005). The ability of an organism to become entrained to external stimuli, mainly light (photic entrainment), is critical for their evolutionary fitness and, often, their health since light/dark cycles on Earth change according to the time of year (Schmal et al., 2020). To be sensitive to changes in light, animals have photoreceptors that begin an entrainment pathway. These can be located in visual organs, nonvisual organs, or both (Tork, 2019). We can plot this conveniently with an actogram  

##### To determine whether entrainment has been effective, we can use the phase angle of entrainment, i.e., the relationship between the timing of the endogenous biological clock and the timing of an external cue (in this case, light/dark) (Pittendrigh, Daan 1976). If an organism is entrained, we would expect the period of locomotor activity to begin close to the time that the light stimulus is applied, and for the period of locomotor activity to stop close to the time of darkness. To measure this, I am computing the average crossings of the beam during the light phase (1) and dark phase (0). By determining the ratio between these two, we can determine whether the spider is entrained - displaying a great enough difference between the light and dark periods to be considered entrained. 

##### Ask about measuring phase angle of entrainment - implies a wave?

```python
  def entrainment(data, column):
    name = column
    dflight=data[data['lights'] == 1]
    dflight=dflight[column]
    
    dfdark=data[data['lights'] == 0]
    dfdark=dfdark[column]
    
    lightmean=np.mean(dflight)
    darkmean=np.mean(dfdark)
    diff=lightmean/darkmean
    
    plt.title('Ratio of Mean Crossings in the Light and Darkness of Spider ' + name)
    y = np.array([lightmean, darkmean])
    plt.pie(y, labels = ['Light', 'Darkness'], shadow = True, explode = [0.2, 0], colors = ['y', '#808080'], normalize = True)
    plt.show()
    
    if diff>.25:
        print('True')
        return True 
    else:
        print('False')
        return False
```
